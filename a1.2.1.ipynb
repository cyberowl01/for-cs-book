## Representation of integers in binary

Convert the following binary (base-2) numbers to decimal (base-10):
1 11001010, 4 00011110,
2 01101101, 5 11100001,
3 1011001,

Convert the following decimal (base-10) numbers to binary (base-2):
1 20,
2 87,
3 123,
4 199,
5 250

### a binary-to-decimal and decimal-to-binary application

```python
def bin2dec(bin_string):
    str_size = len(bin_string)
    rev_str = bin_string[::-1]
    num = 0
    for i in range(str_size):
        b = 2**i
        c=int(rev_str[i])
        x = c*b
        num = num + x
        # print(f"{i} {c} {b} {x}")
    return num

print(bin2dec('10111111'))
print(bin2dec('11001010'))
```
